### Merge sort

### 코드 구현 순서  
1. 길이가 r인 리스트 A를 p~q-1, q~r까지 2개의 리스트로 나누기 위해 개수파악  
2. 리스트를 2개로 나눈 뒤 마지막 원소는 센티넬용 자리를 하나씩 추가해서 많듦.  
    센티넬은 마지막 자리 뒤를 뜻하는 표식(파이썬에는 굳이 없어도 됨)  
3. 두개의 리스트의 같은 자리를 비교하면서 L리스트의 원소가 R리스트 원소보다 작거나 같으면 최초리스트A의 원소와 바꿈  
4. L리스트의 원소가 크면 R리스트 원소를 최초 리스트 원소와 바꿈  

```python
def merge_sort (A,p,q,r):
    n1=q-p+1 # 끝(q)이 포함이라 +1 해야함. 2,3,4 의 개수는 4-2가 아니라 4-2+1
    n2=r-q   # q~r이 4,5,6 이라면 6-4 해서 5,6 총 2개의 원소 개수 확인

    # L[0..n1] , R[0..n2] (마지막 원소가 센티넬)
    # 모든 원소가 0이고 길이는 n1+1, n2+1인 리스트 미리 생성
    L = [0] * (n1 + 1)  # 마지막 1칸은 센티넬용
    R = [0] * (n2 + 1)

    for i in range(n1):
        L[i] = A[p + i]
    for j in range(n2):
        R[j] = A[q + 1 + j]

    L[n1] = float("inf") # 마지막 자리에 센티넬 역할을 하는 float("inf") 추가
    R[n2] = float("inf")

    """ 
    파이썬에서는 미리 리스트 안만들고 슬라이싱해도 됨
    지금은 알고리즘 공부를 위해 센티넬 넣은 코드로 공부!! -파이썬에서는 없어도 코드 구현 가능
    L = A[p:q+1] + [float("inf")]
    R = A[q+1:r+1] + [float("inf")]
    """

    # 파이썬은 0부터 시작하니까 i와 j는 0 할당(CLRS는 1부터 시작)
    # i & j는 L & R 리스트에서 원소를 가져올 위치를 알려주는 포인터 역할
    i = 0
    j = 0
    
    # 리스트 A의 처음부터 끝까지 돌면서 숫자를 배열
    # 두개의 리스트의 같은 자리를 비교하면서 L리스트의 원소가 R리스트 원소보다 작거나 같으면 최초리스트A의 원소와 바꿈  
    # L리스트의 원소가 크면 R리스트 원소를 최초 리스트 원소와 바꿈

    for k in range(p, r + 1):
        if L[i] <= R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1

# 사용 예시
A = [2, 4, 5, 7, 1, 2, 3, 6]   
merge_sort(A, 0, 3, 7)
print(A)  
```

```python
# 주석 삭제한 코드

def merge(A, p, q, r):
    n1 = q - p + 1
    n2 = r - q

    L = [0] * (n1 + 1)
    R = [0] * (n2 + 1)

    for i in range(n1):
        L[i] = A[p + i]
    for j in range(n2):
        R[j] = A[q + 1 + j]

    L[n1] = float("inf")
    R[n2] = float("inf")

    i = 0
    j = 0
    for k in range(p, r + 1):
        if L[i] <= R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1


A = [2, 4, 5, 7, 1, 2, 3, 6]
merge(A, 0, 3, 7)
print(A)
```

```python
#센티넬 없는 버전

def merge_no_sentinel(A, p, q, r):
    L = A[p:q+1]
    R = A[q+1:r+1]

    i = j = 0
    k = p

    # 둘 다 남아있는 동안만 비교
    while i < len(L) and j < len(R):
        if L[i] <= R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1
        k += 1

    # 남은 꼬리 붙이기
    if i < len(L):
        A[k:r+1] = L[i:]
    else:
        A[k:r+1] = R[j:]
```
