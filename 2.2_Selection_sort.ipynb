```python
#  insert sort
def insertion_sort(A):
    for j in range(1, len(A)):      
        key = A[j]                  
        i = j - 1                   
        while i >= 0 and A[i] > key:  
            A[i + 1] = A[i]         
            i -= 1            #i=i-1                  
        A[i + 1] = key              
    return A
```

```python
A = [5, 2, 4, 6, 1, 3]

def selection_sort(A):
    for j in range(len(A)):      
        key = j                           
        for i in range(j+1,len(A)):
            if A[i] < A[key]:
                key=i         
        A[j],A[key]=A[key],A[j]                
    return A

selection_sort(A)
```
