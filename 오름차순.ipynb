for 문의 range 를 벗어났기 때문에 범위 다시 설정 필요

```python
# i 가 i+1 보다 큰 숫자만 리스트 추가
a=[]
for i in range(len(nums)-1):    
    if nums[i]>nums[i+1]:
        a.append(nums[i])

a
```

```python
# i 보다 i+1 숫자가 크면 위치 바꾸기
a=nums.copy()
for i in range(len(a)-1):    
    if a[i]>a[i+1]:
        a[i], a[i+1]=a[i+1], a[i]

a
```

위의 코드는 한번만 루프가 돌기 때문에 이 루프를 반복하는 반복문이 하나 더 필요함.

```python
a=nums.copy()
for i in range(len(a)-1):
    for i in range(len(a)-1):    
        if a[i]>a[i+1]:
            a[i], a[i+1]=a[i+1], a[i]

a
```

for i in range(n - 1 - pass_idx):에서 pass_idx만큼 빼는 이유는, 바깥 루프(pass)를 한 번 돌 때마다 리스트의 맨 뒤쪽 원소들이 “확정적으로 정렬 완료”되기 때문입니다.

버블정렬의 성질은 이렇습니다.

한 번의 pass(안쪽 for문을 처음부터 끝까지 수행)를 마치면,

가장 큰 값이 맨 끝 인덱스(n-1)로 이동해서 고정됩니다.

두 번째 pass를 마치면,

두 번째로 큰 값이 맨 끝에서 두 번째(n-2)로 고정됩니다.

즉 pass_idx가 0, 1, 2… 증가할수록,

뒤에서부터 pass_idx개 원소는 이미 제자리에 있어 더 이상 비교/교환할 필요가 없습니다.

그래서 안쪽 루프의 비교 범위를 줄여:

pass_idx = 0일 때: range(n-1) → i가 0~n-2, 끝까지 비교 (맨 끝 값 확정)

pass_idx = 1일 때: range(n-2) → i가 0~n-3, 마지막 원소는 확정이므로 제외

pass_idx = 2일 때: range(n-3) → i가 0~n-4, 뒤 2개 제외

또한 **n-1**이 들어가는 이유는 항상 a[i]와 a[i+1]를 비교하므로, i의 최대값이 n-2여야 i+1이 n-1을 넘지 않기 때문입니다.

정리하면:

n - 1 : i+1 접근을 위해 필요한 기본 상한

- pass_idx : 매 pass마다 뒤쪽 정렬 완료 구간을 제외하여 불필요한 비교를 줄이는 최적화

빼지 않고 range(n-1)로 계속 돌려도 정렬은 되지만, 이미 확정된 뒤쪽까지 계속 비교해서 비효율적입니다.

```python
# 기존 루프에서 i 번째만큼을 빼서 내부루프로 들어감. range(len(a)-1-i)

a=nums.copy()
for i in range(len(a)-1):
    for i in range(len(a)-1-i):    
        if a[i]>a[i+1]:
            a[i], a[i+1]=a[i+1], a[i]

a
```
