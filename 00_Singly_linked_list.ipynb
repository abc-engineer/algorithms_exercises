### 특징

원형큐의 경우 크기가 고정크기로 제한되기 때문에 실무에서 해결 못하는 문제들이 발생.
이때 사용되는 자료구조가 연결리스트.
- 원소가 없으면 0, 원소를 추가하면 추가하는 만큼 크기가 n으로 커짐.
- 연결리스트로 큐를 구현하면 연결 리스크의 크기를 자유롭게 변경 가능
- 단일 연결 리스트 simply linked list와 이중 연결 리스트 doubly linked list 등이 있음
- C계열 처럼 진짜 주소 값을 저장하는 경우는 next에 주소값을 넣지만 파이썬은 실제 주소값 말고 객체 참조(reference)를 저장


#### 구성
연결리스트의 한칸(노드)은 데이터 + 넥스트(다음 노드 주소)로 구성 됨
- 데이터: 실제값
- 넥스트: 다음 노드가 어디있는지 알려주는 기능  

노드  
- [data | next→]  
- [10 | 다음→] 

연결리스트    
- [data | next→] → [data | next→] → [data | next→] → 끝(nil/None)  
- [10 | 다음→] → [20 | 다음→] → [30 | 다음→] → 끝(nil/None)



### 장단점 
장점
- 중간 삽입/삭제가 비교적 쉬움
- 배열은 중간에 넣거나 빼면 뒤에 있는 것들을 줄줄이 “밀기/당기기” 해야 함
- 연결 리스트는 “연결(주소)만 갈아끼우면” 됨  
- 예를 들어 
    10 → 20 → 30 사이에 25를 넣고 싶다면,  
    **원래:**  
    20이 30을 가리킴  
    **바꾼 뒤:**  
    20이 25를 가리키게 하고  
    25가 30을 가리키게 하면 끝  
    즉, “주소 두 개만 바꾸면” 중간 삽입  

단점
- n번째 요소 접근이 느림(순차로 따라가야 함)
- 노드마다 next 저장공간이 추가로 필요(메모리 오버헤드)

### 리스트(배열)과 연결리스트의 차이
리스트(배열)
- 데이터가 메모리에서 붙어서 저장되는 느낌
- 인덱스로 바로 접근: a[5] 같은 게 빠름

연결 리스트
- 데이터가 메모리 여기저기에 흩어져 있어도 됨
- 대신 “다음 칸 주소”를 따라가며 찾아감  
 5번째 노드를 보려면 1→2→3→4→5 순서로 걸어가야 함

### 작동(동작방식)

연결 리스트의 첫번째로드는 Head 노드  
head  
&nbsp;&nbsp;↓  
[10 | next] → [20 | next] → [30 | next] → nil/None  

이 head 노드를 유지하면 알고리즘을 간단하게 구현할 수 있음, (필수는 아님.)

#### 노드 추가하기
1. head 노드만 있는 리스트 생성
2. 노드에 추가할 값 복사 후 next에 nil/None 할당
3. 새 노드 생성 후 2번째 추가할 값 복사 next에 첫번째 노드가 참조 되게 객체 참조 또는 포인터(파이썬에서는 가상 개념, C언어 개념) 연결

#### 노드 순회하기
1. head node에 데이터를 넣을지 더미 head (비어있는 head)로 둘지 결정  
    더미 헤드를 사용할 경우 첫번째 노드가 삭제되거나 추가 되어도 새로 갱신할 필요 없음  
    다만 첫 헤드를 비울 경우 메모리 낭비는 발생  
2. head에 값이 있으면 head를, 더미 head의 경우 head.next가 가르키는 cur(cursor)이라는 변수명을 지닌 객체 준비
3. cur이 가리키는 노드의 값을 출력(또는 사용)
4. 다음 cur이 next를 가리키도록 변경 (cur.next)
5. cur.next가 nil/None을 가리키면 순회 종료

#### 노드 삭제하기
코드 단순화를 위헤 head는 값이 없는 더미 head로 정리
1. 더미 head node를 가리키는 객체 pre와 head.next를 가리키는 cur 객체 준비 (편의상 두개의 포인터를 준비한다고 이해)
2. 찾고자 하는 target 값을 찾을때 까지 cur 포인터가 이동(target 값이 아닌 순서로도 찾을 수 있음)
3. target 값은 고유값이라고 가정 후 진행
4. 이때 pre포인터도 cur을 따라 한칸씩 이동
5. cur가 target 값을 찾으면 cur가 있는 노드의 next값을 pre 노드 next에 할당 후 해당 노드 삭제
6. cur은 해당 노드가 삭제 되며 앞의 노드로 이동
