```python
import re
n = 6542
str_n=[]

str1=re.findall(r"\d",str(n)) # 숫자를 str으로 변환 후 추출

for i in str1: #다시 숫자로 변환
    str_n.append(int(i))

result= sum(str_n) #리스트 요소들의 합을 구합

print(result)
```

```python
import re
n = 521
str_n=[]
result=1
str1=re.findall(r"\d",str(n)) # 숫자를 str으로 변환 후 추출

for i in str1: #다시 숫자로 변환
    str_n.append(int(i))

for j in str_n: # 숫자로 변환된 리스트의 곱을 구함.
    result=result*j

print(result)
```
