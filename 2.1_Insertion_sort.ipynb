### 2.1 Instance 정렬 문제



INSERTION-SORT(A)  
1  for j = 2 to A.length  
2      key = A[j]  
3      // Insert A[j] into the sorted sequence A[1..j-1].  
4      i = j - 1  
5      while i > 0 and A[i] > key  
6          A[i + 1] = A[i]  
7          i = i - 1  
8      A[i + 1] = key  

```python
# 파이썬 코드로 변환
def insertion_sort(A):
    for j in range(1, len(A)):      
        key = A[j]                  
        i = j - 1                   
        while i >= 0 and A[i] > key:  
            A[i + 1] = A[i]         
            i -= 1
            #i=i-1                  
        A[i + 1] = key              
    return A
```

```python
A = [5, 2, 4, 6, 1, 3]
insertion_sort(A)
print(A)
```

### 코드 설명  
리스트의 2번째 자리에 있는 수를 기준(key)로 설정  
i=1부터 for문 loop 실행  
while문에 사용할 i는 j-1부터 loop  
Key 값보다 A[i]값이 크다면 A[i+1]값을 A[i]로 바꿈  
무한루프가 되지 않도록 i값은 -1씩 줄여가며 key값과 비교하는 while loop를 실행  
while loop가 종료 되면 A[i+1]위치에 key 값을 넣어줌(바뀐숫자 앞에 키값이 들어감)
