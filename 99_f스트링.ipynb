```python

# f-string은 문자열 앞에 F를 붙이고, 중괄호 {}를 사용해 변수나 표현식을 삽입
# 문자열 안에서 변수 값을 그대로 출력할 수 있어 코드가 간결
# 띄어쓰기도 됨
# 변수간 타입 안맞춰도 사용 가능
name = "난개발"
age = 29
print(f"내 이름은 {name} 이고, {age}살 이야.")

# 중괄호 안에서 연산도 가능
a = 5
b = 3
print(f"{a} 와 {b}의 합은 {a + b}.")

```

```python
num1 = 12345
num2 = 45
num3 = 3.141592
print(f"Number: {num1}")     # 그냥 출력
print(f"Number: {num2:5d}")  # 총 5자리로 정렬
print(f"Number: {num3:.3f}") # 소수점 셋째 자리까지 출력
```

```python
print(f"Number: {num3:5d,.2f}") # 소수점 둘째 자리까지 출력
```

```python
num = 10
print(f"Number: {num:>10}")  # 그냥 출력
print(f"Number: {num:<10}")  # 그냥 출력
print(f"Number: {num:^10}")  # 그냥 출력
```
