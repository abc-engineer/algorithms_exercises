```python
def fibonacci(n):
    """n번째까지의 피보나치 수열을 계산해 리스트로 반환."""
    fib = [0] * (n + 1)  # 피보나치 수열 저장 리스트
    fib[1] = 1  # F₁ = 1
    if n > 1:
        fib[2] = 1  # F₂ = 1

    for i in range(3, n + 1):
        fib[i] = fib[i - 1] + fib[i - 2]  # 피보나치 수열 계산

    return fib

def count_divisible_pairs(S):
    """집합 S에 대해 (i, j) 쌍에서 Fi가 Fj로 나누어 떨어지는 개수를 반환."""
    max_index = max(S)  # 주어진 집합 S에서 가장 큰 인덱스를 찾음
    fib = fibonacci(max_index)  # 피보나치 수열 계산

    count = 0  # 조건을 만족하는 순서쌍 개수
    for i in S:
        for j in S:
            if i != j and fib[i] % fib[j] == 0:  # 나누어 떨어지는지 확인
                count += 1

    return count

# 예제 입력
S = list(range(1,10))  # 집합 S
result = count_divisible_pairs(S)
print(result)  # 조건을 만족하는 순서쌍 개수 출력
```
