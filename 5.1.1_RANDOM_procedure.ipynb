### Random procedure

#### 문제: RANDOM(0,1)만 호출해 동작하도록 RAMDOM(a,b)의 구현 코드 작성

0~4 까지 5개의 숫자를 공정하게 뽑으려면? 
1. RANDOM(0, 1)을 이용해서 비트를 3번 호출한다.
    - 1번째: 0
    - 2번째: 1
    - 3번째: 0
    그러면 비트열이 010 
    이렇게 3번 던지면 나올 수 있는 비트열은 총 8가지  
    000, 001, 010, 011, 100, 101, 110, 111  
    그리고 공정한 동전이면 이 8가지는 각각 확률 1/8  
2. 이때 각각의 2진수에 해당하는 정수를 맵핑

    | 비트열 | 10진수 |
    | --- | ---- |
    | 000 | 0    |
    | 001 | 1    |
    | 010 | 2    |
    | 011 | 3    |
    | 100 | 4    |
    | 101 | 5    |
    | 110 | 6    |
    | 111 | 7    |

    즉, 동전 3번 던지기 = 0~7 중 하나를 1/8로 고르는 것과 같음  

3. 목표는 0,1,2,3,4 를 각각 1/5로 뽑는 것.
    - 그런데 우리는 0~4 (5개)만 필요함
    - 하지만 지금은 0~7(8개)을 균등하게 뽑음.
    - 그래서 8개 중에서 “필요한 5개”만 쓰고, 나머지 3개는 버림.
        - 채택(accept): 0,1,2,3,4
        - 거부(reject): 5,6,7
    - x가 0~4면 그 값을 그대로 결과로 사용
    - x가 5~7이면 그건 버리고 처음부터 다시
    - 예를 들어:
        - 비트열 011 → x=3 → 0~4 안에 있음 → 결과 3 (끝)
        - 비트열 110 → x=6 → 5~7임 → 버림 → 다시 3비트 생성

4. 확률 계산
    한 번 뽑을 때:

    - 채택 확률: P(x<5)=5/8    
    - 거부 확률: 3/8
        

    이제 특정 값 $j \in \{0,1,2,3,4\}$가 최종 결과가 될 확률을 계산해보면,

    - 첫 시도에서 바로 j가 나올 확률: $1/8$    
    - 한 번 거부(5,6,7 중 하나)되고 다음에 j가 나올 확률: $(3/8)\cdot(1/8)$  
    - 두 번 거부되고 그다음 $j: (3/8)^2\cdot(1/8)$
    - …
        
    그래서 총확률은 기하급수 합으로:

    $\displaystyle P(\text{결과}=j)=\frac{1}{8}\left(1+\frac{3}{8}+\left(\frac{3}{8}\right)^2+\cdots\right)$  
    $\;\;\;\;\displaystyle=\frac{1}{8}\cdot \frac{1}{1-\frac{3}{8}} =\frac{1}{8}\cdot \frac{1}{\frac{5}{8}} =\frac{1}{5}$

    모든 j=0,1,2,3,4에 대해 똑같이 1/5가 나오죠.  
    즉 **편향이 사라집니다.**

    > 핵심: “채택 구간(0~4)” 안에서는 각 값이 **원래도 동일 확률(1/8)** 이었기 때문에,  
    > 조건부로 봐도 균등이 유지됩니다.



#### 의사코드: RANDOM(a, b) using RANDOM(0, 1)  
- 아래는 RANDOM(0,1)(공정한 0/1) 만 호출해서 RANDOM(a,b)(a~b 균등) 를 만드는 표준 구현.
- 핵심은 비트로 수를 만들고(representation), 범위를 넘으면 버리는(rejection sampling) 방식.

```text
RANDOM(a, b)
1  n = b - a + 1
2  k = ceil(lg n)                 // lg는 log2 ceil은 천장 함수(ceiling function), 실수𝑥를 올림해서 정수로
3  repeat
4      x = 0
5      for i = 1 to k
6          x = 2*x + RANDOM(0,1)  // k비트 정수 생성: 0 .. 2^k - 1
7  until x < n
8  return a + x
```
