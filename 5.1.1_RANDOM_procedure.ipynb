### Random procedure

의사코드: RANDOM(a, b) using RANDOM(0, 1)  
- 아래는 RANDOM(0,1)(공정한 0/1) 만 호출해서 RANDOM(a,b)(a~b 균등) 를 만드는 표준 구현.
- 핵심은 비트로 수를 만들고(representation), 범위를 넘으면 버리는(rejection sampling) 방식.

```text
RANDOM(a, b)
1  n = b - a + 1
2  k = ceil(lg n)                 // lg는 log2
3  repeat
4      x = 0
5      for i = 1 to k
6          x = 2*x + RANDOM(0,1)  // k비트 정수 생성: 0 .. 2^k - 1
7  until x < n
8  return a + x
```
