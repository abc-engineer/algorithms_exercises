```python
import re

text = "예제 문장: 1,000.5억 달성, 300억 투자, 12,345억 가치, 5,000만 기부, 100만 달성"

# 정규식 패턴: 숫자와 억 또는 만 단위를 포함한 전체 패턴 추출
pattern = r'\b\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:억|만)\b'

# 전체 숫자와 단위를 묶어서 추출
matches = re.findall(pattern, text)

# 억과 만을 숫자로 환산해서 계산하는 함수
def convert_to_number(value):
    # '억' 단위가 포함된 경우
    if '억' in value:
        number = float(value.replace('억', '').replace(',', ''))
        return number * 10**8
    # '만' 단위가 포함된 경우
    elif '만' in value:
        number = float(value.replace('만', '').replace(',', ''))
        return number * 10**4
    return 0

# 추출된 값들을 숫자로 환산
converted_numbers = [convert_to_number(match) for match in matches]

print(converted_numbers)
```

```python
import re

text = "예제 문장: 1,000.5억 달성, 300억 투자, 12,345억 가치, 5,000만 기부, 100만 달성"

# 정규식 패턴: 숫자와 억 또는 만 단위를 포함한 전체 패턴 추출
pattern = r'\b\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:억|만)\b'

# 전체 숫자와 단위를 묶어서 추출
matches = re.findall(pattern, text)

# 억과 만을 숫자로 환산해서 계산하는 함수
def convert_to_number(value):
    # '억' 단위가 포함된 경우
    if '억' in value:
        number = float(value.replace('억', '').replace(',', ''))
        return int(number * 10**8)  # 정수로 변환
    # '만' 단위가 포함된 경우
    elif '만' in value:
        number = float(value.replace('만', '').replace(',', ''))
        return int(number * 10**4)  # 정수로 변환
    return 0

# 추출된 값들을 숫자로 환산 후 정수로 변환
converted_numbers = [convert_to_number(match) for match in matches]

print(converted_numbers)
```

```python
import re

text = "예제 문장: 1,000.5억 달성, 300억 투자, 12,345억 가치, 5,000만 기부, 100만 달성"

# 정규식 패턴: 숫자와 억 또는 만 단위를 포함한 전체 패턴 추출
pattern = r'\b\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:억|만)\b'

# 전체 숫자와 단위를 묶어서 추출
matches = re.findall(pattern, text)

# 억과 만을 숫자로 환산하고 백만원 단위로 계산하는 함수
def convert_to_number(value):
    # '억' 단위가 포함된 경우
    if '억' in value:
        number = float(value.replace('억', '').replace(',', ''))
        result = int(number * 10**8) // 10**6  # 백만원 단위로 변환
        return "{:,} 백만원".format(result)  # 3자리마다 쉼표 추가
    # '만' 단위가 포함된 경우
    elif '만' in value:
        number = float(value.replace('만', '').replace(',', ''))
        result = int(number * 10**4) // 10**6  # 백만원 단위로 변환
        return "{:,} 백만원".format(result)  # 3자리마다 쉼표 추가

# 추출된 값들을 숫자로 환산 후 백만원 단위로 변환
converted_numbers = [convert_to_number(match) for match in matches]

print(converted_numbers)
```
