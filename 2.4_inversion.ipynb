### Inversion

#### 정의
> $A[1:n]$을 $n$ 개의 서로다른 원소의 배열이라 했을 때, $i<j$이고 $A[i] <A[j]$라 하면 $(i,j)$ 쌍을 역위라 함.

#### 코드 구현 순서  
1. i와 i보다 큰수 j를 변수로 할당  
2. A[i]>A[j] 이면 그 숫자를 카운트

```python
def inversion(A):
    cnt=0
    for i in range(len(A)):
        for j in range(i+1,len(A)):
            if i < j and A[i] > A[j]:
                cnt +=1
    return print(f"number of inversion: {cnt}")
```

```python
A=[2,4,1,3,5]
inversion(A)
```

최대 역위 수는 $\displaystyle\frac{n(n-1)}{2}$​ (모든 쌍이 뒤집힘)  
- 배열이 내림차순이면 모든 순서쌍이 역위가 됨.  
- 최대 역위는 내림차순에서 n개의 원소중 서로다른 2개를 고르는 방법이기 때문에 $\displaystyle\frac{n(n-1)}{2}$​개  

```python
B=[5,4,3,2,1]
inversion(B)
```
