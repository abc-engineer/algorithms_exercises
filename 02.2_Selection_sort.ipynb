### Selection sort

선택 정렬은 버블이나 삽입정렬과 다르게 먼저 최소값을 찾아 맨 앞으로 가져와야 함.  
그 다음에 2번째 자리에서 다시 최소값을 찾아 1번째 자리로 이동하는 loop실행  
최소값과 맨 앞에 있는 값의 자리를 바꿔야 함. (버블 처럼)

```python
# 참고용 insert sort
def insertion_sort(A):
    for j in range(1, len(A)):      
        key = A[j]                  
        i = j - 1                   
        while i >= 0 and A[i] > key:  
            A[i + 1] = A[i]         
            i -= 1            #i=i-1                  
        A[i + 1] = key              
    return A
```

```python
A = [5, 2, 4, 6, 1, 3]

def selection_sort(A):
    for j in range(len(A)):      
        key = j                           
        for i in range(j+1,len(A)):
            if A[i] < A[key]:
                key=i         
        A[j],A[key]=A[key],A[j]                
    return A

selection_sort(A)
```

```python
# 맨마지막은 자리가 len(A)번째에 바뀌어 있기 때문에 실행 불필요 len(A)-1 까지만 반복

A = [5, 2, 4, 6, 1, 3]

def selection_sort(A):
    for j in range(len(A)-1):      
        key = j                           
        for i in range(j+1,len(A)):
            if A[i] < A[key]:
                key=i         
        A[j],A[key]=A[key],A[j]                
    return A

selection_sort(A)
```
