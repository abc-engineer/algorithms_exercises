정규표현식 또는 정규식 이란 것을
공부하고 전에 풀다가 포기한 문제를
풀 수 있게 되었다. 

문제는 간단했다.
임의의 숫자의 각 자리수의
합을 구해라.

if문에 elif에 말도 안되게 난리치며
작성하다가 포기했는데 
정규식을 이용해서
숫자만 추출해서 계산 하니 넘 쉬움

숫자만 추출하는 것은 "\d"
re.findall(r"\d",str(n))
하면 텍스트로 전환된 숫자만 추출가능

이걸 다시 숫자로 변환해서
리스트로 만들고 그걸 더 함.

그런데 이것도 고수들은 한 줄 컷이겠지?

1|i I L 0 o O ㅇ

```python
import re
n = 6542
str_n=[]

str1=re.findall(r"\d",str(n)) # 숫자를 str으로 변환 후 추출

for i in str1: #다시 숫자로 변환
    str_n.append(int(i))

result= sum(str_n) #리스트 요소들의 합을 구합

print(result)
```

```python
import re
n = 521
str_n=[]
result=1
str1=re.findall(r"\d",str(n)) # 숫자를 str으로 변환 후 추출

for i in str1: #다시 숫자로 변환
    str_n.append(int(i))

for j in str_n: # 숫자로 변환된 리스트의 곱을 구함.
    result=result*j

print(result)
```
