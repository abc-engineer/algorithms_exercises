### 문제

0과 1을 각각 1/2의 확률로 출력하고자 한다.  
함수 BIASED-RANDOM은 0 < p < 1인 p의 확률로 1을 출력하고 1 - p의 확률로 0을 출력하지만, p 값은 모른다.  
BIASED-RANDOM을 서브루틴으로 사용해 0과 1을 똑같이 1/2의 확률로 리턴하는 알고리즘을 제시하라. 

> 0과 1을 1/2 확률로 리턴하는 함수를 만들어라!

#### 아이디어

1. von Neumann 트릭
2. BIASED-RANDOM()을 두 번씩 호출해서 결과가 서로 다를 때만 그 쌍을 채택하면 확률이 1/2
    - (0,1)이 나올 확률: $(1-p)\cdot p$    
    - (1,0)이 나올 확률: $p\cdot(1-p)$  
    두 확률이 **항상 동일**하므로,
    - (0,1)이면 0을 반환    
    - (1,0)이면 1을 반환    
    으로 정하면 0과 1이 **각각 1/2 확률** 
    (0,0)이나 (1,1)은 버리고 다시 시도

#### 의사코드
```text
UNBIASED-RANDOM()
1  while TRUE
2      a ← BIASED-RANDOM()
3      b ← BIASED-RANDOM()
4      if a = 0 and b = 1
5          return 0
6      else if a = 1 and b = 0
7          return 1
8      // else (a = b): ignore and repeat

```

#### Flow
1. 1일 확률 p, 0일 확률을 1−p로 반환하는 “편향 동전” 서브루틴 구성 (BIASED-RANDOM)
    - p는 0<p<1인 어떤 고정값(임의의 값 p)
2. 이때 BIASED-RANDOM 함수를 2번 호출해서 각각의 a,b 반환
    - 단, 임의의 확률 p값은 a,b 동일
3. while TRUE loop를 통해 a와 b가 다른 값일때 0 또는 1을 반환
4. 함수를 호출할때는 실행시 마다 값이 바뀌어야 하니까 lambda사용

````python
import random

def biased_random(p: float) -> int:
    """
    BIASED-RANDOM(): p의 확률로 1을 반화 , else 0
    (원래 문제에서는 p는 모르는 값.)
    """
    if not (0.0 < p < 1.0):
        raise ValueError("p must be in (0, 1)")
    return 1 if random.random() < p else 0

def unbiased_random(biased_fn) -> int:
    """
    unbiased_random(): biased_random 함수를 이용해서 0과 1을 각각 1/2의 확률로 리턴
    """

    while True:
        a = biased_fn()
        b = biased_fn()

        if a == 0 and b == 1:
            return 0
        elif a == 1 and b == 0:
            return 1
        else:
            continue
````

````python
x=lambda: biased_random(0.8)# 호출할 때마다 biased_random(0.5) 실행
# x = biased_random(0.5)로 변수를 저장하면 이미 0이나 1이 x에 할당되기 때문에 다시 던지는 역할 을 할 수 없음
#p=0.8이면 (00,11)도 자주 나와서(특히 11이 많음) 반복이 많음
unbiased_random(x)
````

````python
# 람다를 안쓰고 사용자 정의 함수로 처리 할 수 있음
def biased_fn():
    return biased_random(0.5)

unbiased_random(biased_fn)
````

````python
N = 100000
ones = sum(unbiased_random(x) for _ in range(N))
print(ones / N)   # 대략 0.5 근처
````

### 파이썬스럽게 코드 구성
a 와 b가 다르면 그냥 a만 리턴해도 같은 값

````python
def biased_random_py(p: float) -> int:
    if not (0.0 < p < 1.0):
        raise ValueError("p must be in (0, 1)")
    return 1 if random.random() < p else 0

def unbiased_random_py(biased_fn):
    while True:
        a = biased_fn()
        b = biased_fn()
        if a != b:
            return a
````

````python
y=lambda: biased_random_py(0.5)
unbiased_random_py(y)
````
