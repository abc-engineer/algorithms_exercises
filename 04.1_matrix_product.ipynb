### matrix_product 점곱(dot product)



$ A=\begin{pmatrix} \ a_{11}&a_{12}\\ a_{21}&a_{22}\end{pmatrix}, B=\begin{pmatrix} \ b_{11}&b_{12}\\ b_{21}&b_{22}\end{pmatrix}$ 일때,  

$ A \cdot B= C \begin{pmatrix} \ c_{11}&c_{12}\\ c_{21}&c_{22}\end{pmatrix}$ 를 구하기

matrix C 계산

$c_{11}=a_{11}\cdot b_{11}+a_{12}\cdot b_{21}$  
$c_{12}=a_{11}\cdot b_{12}+a_{12}\cdot b_{22}$  
$c_{21}=a_{21}\cdot b_{11}+a_{22}\cdot b_{21}$  
$c_{22}=a_{21}\cdot b_{12}+a_{22}\cdot b_{22}$  

### flow

1. matrix A,B에 대해 $n\times n$ 형태 확인  
    $n\times n$ 0 행렬 C를 만들어야 함.
2. $\displaystyle c_{ij}=\sum_{k=1}^n a_{ik} \cdot b_{kj}$ 수식 구현
3. 반복되는 변수 $i,j,k$에 대해 어떤 변수가 먼저 반복 되어야 하는지 확인  
    k -> j -> i 순서로 숫자가 상승하기 때문에 k가 가장 안쪽에서 반복 되야함.
4. $\displaystyle \sum_{k=1}^n$ 을 구현하기 위해 $c_{ij}$에 $a_{ik}\cdot b_{kj}$를 더하는 수식을 구현해야함.  
5. 사용자 정의 함수에서 C를 리턴해야 함.

```python
def matrix_product(A, B, n):
    C = [[0]*n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j]=C[i][j]+A[i][k]*B[k][j]
    return C
```

```python
A = [
    [1, 2],
    [3, 4]
]
B = [
    [5, 6],
    [7, 8]
]
```

```python
matrix_product(A, B, 2)
```
