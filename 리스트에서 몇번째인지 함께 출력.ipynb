```python
a=['Sun','Mon', 'Tue', 'Wed', 'Thu','Fri','Sat']
```

```python
# 리스트내의 인자만 출력

for el in a :
    print(el)
```

```python
# 리스트 번호만 출력

for el in range(len(a)) :
    print(el)
```

```python
# range(leg()) 함수를 쓰면서 인자를 불러오고 싶을때

for el in range(len(a)) :
    print(a[el])
```

```python
# 리스트 번호가 0부터 출력되는게 싫으때

for i in range(len(a)) :
    print(i+1)
```

```python
# 두개 한번에 출력하기
# 숫자와 문자는 함께 출력할 수 없으니까 str 함수로 숫자는 문자로 변환

for i in range(len(a)) :
    print(str(i+1) +' ' + a[i])
```

```python
#간단하게 enumerate함수로 함께 출력

for i, el in enumerate(a) :
    print(i, el)
```

```python
#간단하게 enumerate함수로 인덱스와 요소 함께 출력
#0부터 카운팅하기 싫을때

a=['Sun','Mon', 'Tue', 'Wed', 'Thu','Fri','Sat']
for i, el in enumerate(a) :
    print(i+1, el)
```
