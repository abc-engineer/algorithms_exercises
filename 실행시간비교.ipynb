Comparison of running times For each function f.n/ and time t in the following table, determine the largest size n of a problem that can be solved in time t, assuming that the algorithm to solve the problem takes f.n/ microseconds.  

실행 시간 비교(Comparison of running times)
아래 표에 주어진 각 함수 f(n)과 시간 t에 대해, 문제를 t 시간 안에 해결할 수 있는 입력 크기 
n의 최댓값을 구하시오. 단, 그 문제를 해결하는 알고리즘의 실행 시간은 )
f(n) 마이크로초(μs)가 걸린다고 가정한다.



```python
# 시간 단위 → 초(seconds)
SECONDS = {
    "second": 1,
    "minute": 60,
    "hour": 60 * 60,
    "day": 24 * 60 * 60,
    "month": 30 * 24 * 60 * 60,     # 30일 가정
    "year": 365 * 24 * 60 * 60,     # 365일 가정
    "century": 100 * 365 * 24 * 60 * 60,
}

MICROSECONDS_PER_SECOND = 10**6
```

```python
def to_microseconds(unit, amount) ->int:
    return amount * SECONDS[unit] * MICROSECONDS_PER_SECOND
```

```python
to_microseconds("minute", 1)
```

```python
def to_microseconds(unit, amount) ->int:
    return amount * SECONDS[unit] * MICROSECONDS_PER_SECOND

to_microseconds("minute")
```

```python
def to_microseconds(unit, amount:int=1) ->int:
    return amount * SECONDS[unit] * MICROSECONDS_PER_SECOND

to_microseconds("minute")
```

```python
#unit 도 어떤 타입인지 힌트를 주면 더 정확한 코드가 완성

def to_microseconds(unit:str, amount:int=1) ->int:
    return amount * SECONDS[unit] * MICROSECONDS_PER_SECOND

to_microseconds("minute")
```

```python
# 예: 표의 t 값(각 열)을 μs로 만들기
t_units = ["second", "minute", "hour", "day", "month", "year", "century"]
t_microseconds = {u: to_microseconds(u) for u in t_units}
```
